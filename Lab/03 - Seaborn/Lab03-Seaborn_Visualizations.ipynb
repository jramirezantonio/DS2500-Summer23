```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')
penguins.dropna(inplace=True)
penguins
```

```python
sns.pairplot(penguins, hue='species')
```

```python
# it looks like there's a strong positive correlation between flipper_length_mm and body_mass_g
# with at least 2 apparent clusters of species; both the Chinstrap and Adelie species have a lower
# body mass than the Gentoo species
# I suspect penguins from the Gentoo species are significantly heavier than the other species, and potentially
# the Gentoo are amongst the heaviest penguins species
```

```python
plt.figure(figsize = (5,5), dpi=500)
sns.scatterplot(x = penguins.flipper_length_mm, y = penguins.body_mass_g, hue = penguins.species)
plt.grid()
plt.title('Flipper Length V.S. Body Mass by Species')
```

```python
# we can see that the lower the flipper length of a penguin, the lower will its body mass be, especially
# for the Adelie and Chinstrap species
```

```python
plt.figure(figsize = (5,5), dpi = 500)
sns.boxplot(data=penguins, x='sex', y='body_mass_g', palette = 'pastel')
sns.swarmplot(data=penguins, x='sex', y='body_mass_g', hue='species')
plt.title('Body Mass by Species Swarmplot')
#plt.grid()
```

```python
# swarmplot is used for categorical data on x-axis (sex) v.s. body mass
# swarmplot avoids overlap of values even though there are only 2 sexes (male, female)
# it's clear that the Gentoo species is heavier than the other 2, and we can also see that
# males are typically heavier than females for all 3 species
# Side observation: it would be hard to distinguish Adelie from the Chinstrap species using a 
# classifier given that they form a cluster, as opposed to the Gentoo species that forms
# a cluster of its own
```
