```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as stats
```

```python
bb = pd.read_csv('brain_body.csv')
bb['body_kg'] = bb.body_gm / 1000

bb
```

```python
plt.figure(figsize=(6,6), dpi=200)

plt.bar(bb.Species, bb.brain_gm)
plt.xticks(rotation='vertical')
#plt.yscale('log')
plt.title("Brain mass (g) by Species")
plt.ylabel("Brain mass (g)")
plt.grid()
plt.show()
```

```python
bb['brain_body_ratio'] = bb.brain_gm / bb.body_kg
bb = bb.sort_values(by='brain_body_ratio', ascending=False)
bb
```

```python
plt.figure(figsize=(6,6), dpi=100)
sns.regplot(x=bb.body_kg, y=bb.brain_gm)
plt.grid()
plt.title('Brain vs. Body Weight')
plt.xscale('log')
plt.yscale('log')
```

```python
bb['log_brain'] = np.log10(bb.brain_gm)
bb['log_body'] = np.log10(bb.body_kg)
```

```python
plt.figure(figsize=(6,6), dpi=100)
sns.regplot(x=bb.log_body, y=bb.log_brain)
plt.grid()
plt.title('Brain vs. Body Weight')
plt.text(math.log10(73)-.1, math.log10(1350)+.1, 'human')
plt.text(math.log10(3.2)+.1, math.log10(375)+.1, 'newborn')
```

```python
lm = stats.linregress(x=bb.log_body, y=bb.log_brain)
print(lm.slope)
print(lm.intercept)
```

```python
bb['pred_brain'] = lm.slope * bb.log_body + lm.intercept
bb['enceph'] = bb.log_brain - bb.pred_brain
bb
```

```python
bb = bb.sort_values(by='enceph', ascending=False)
plt.figure(figsize=(6,6))
plt.bar(bb.Species, bb.enceph)
plt.xticks(rotation="vertical")
plt.ylabel("Ecephalization")
plt.grid()

plt.show()
```
