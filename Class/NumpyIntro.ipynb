```python
A = np.zeros((5, 4), dtype = int)
A
```

```python
A[1,1] = 3
A[1][1] = 3
A
```

```python
A[0,2] = 8
A[3,3] = 5
A[2,0] = 1
A[3,2] = 20
A
```

```python
import matplotlib.pyplot as plt
plt.imshow(A, cmap = 'viridis', interpolation = 'hamming', vmin = 0, vmax = 30) # default colormap is not grey, instead default is viridis
# vmin and vmax change the min and max colors without changing values of the array
# interpolation 'smooths out' image, and there are various options to choose from
A.max()
```
