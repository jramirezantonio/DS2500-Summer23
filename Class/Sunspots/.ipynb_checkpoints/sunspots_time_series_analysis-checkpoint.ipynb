```python
# A demo of time series visualizaton and analysis
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# read data 
sun = pd.read_csv('SN_m_tot_V2.0 (1).csv', delimiter=';', header=None)
sun = sun.iloc[:, 2:4]
sun.columns = ['year', 'spots']
```

```python
sun['mavg'] = sun.spots.rolling(window=11).mean()

plt.figure(figsize=(10, 3), dpi=300)
plt.scatter(sun.year, sun.spots, marker='.', label='observed')
plt.plot(sun.year, sun.mavg, c='r', label='11-year moving average')
plt.legend()
plt.grid()
plt.xlabel('Year')
plt.ylabel('# Sunspots')
plt.axvspan(1790, 1830, color='grey', alpha=0.25)
plt.text(1810, 305, 'Dalton Minimum', ha='center', fontsize=8)

plt.axvspan(1955, 2005, color='grey', alpha=0.25)
plt.text(1980, 380, 'Modern Maximum', ha='center', fontsize=8)

plt.show()
```
