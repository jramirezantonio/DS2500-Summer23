```python
# A demonstration of some image-generation capabilities of numpy
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SIZE = 200

image = np.zeros((SIZE, SIZE), dtype = int)
for i in range(SIZE):
    for j in range(SIZE):
        image[i,j] = (i*i + j*j) % 201
plt.figure(figsize = (10,10))
plt.imshow(image, cmap = 'Spectral', interpolation = 'gaussian')
plt.show()
```

```python
SIZE = 200

image = np.zeros((SIZE, SIZE, 3), dtype = int)
for i in range(SIZE):
    for j in range(SIZE):
        red = (i * j) % 256
        green =(i * i - j*j*i) % 256
        blue = (i*i + j*j) % 256
        
        image[i,j] = [red, green, blue]
#         image[i,j,0] = red
#         image[i,j,1] = green
#         image[i,j,2] = blue
        # image[i,j] = (i*i + j*j) % 201
plt.figure(figsize = (10,10))
plt.imshow(image, cmap = 'Spectral', interpolation = 'None')
plt.show()
```

```python
import matplotlib.image as img

def show_image(image):
    plt.imshow(image)
    plt.show()
    
# crop the image
def crop(image, fr, to):
    '''
    Crop and image
    fr - upper left coordinate (tuple)
    to - lower right coordinate (tuple)
    return the cropped image e.g., (400,0) to (1000,600)
    '''
    # image[400:1000, 0:600]
    cropped = image[fr[0]:to[0], fr[1]:to[1]]
    return cropped

def blur(image, level=10):
    blurry = image.copy()
    for i in range(level, image.shape[0] - level):
        for j in range(level, image.shape[1] - level):
            blurry[i,j] = image[i-level:i+level, j-level:j+level].mean()
    return blurry
            
# Read the image into a numpy array
original = img.imread('asteroid.png')[:,:,0]
cropped = crop(original, (400,0), (1000,600))
blurred = blur(cropped, level = 10)

show_image(original)
show_image(cropped)
show_image(blurred)
```
